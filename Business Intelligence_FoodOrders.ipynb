### Q. Which user_id ordered most items in each city

### Q. What was percent change in revenue for chennai for 13 august compared to 12 august

```python
df
```

```python
chennai_df=df.groupby('city').get_group('Chennai')
chennai_df
```

```python
chennai_sale_df=chennai_df.groupby('order_date').agg({'price':'sum'}).reset_index()
chennai_sale_df
```

```python
chennai_sale_df['Previous_day_sale']=chennai_sale_df['price'].shift(1)
chennai_sale_df
```

```python
chennai_sale_df['revenue_change']=(chennai_sale_df['price']-chennai_sale_df['Previous_day_sale'])/chennai_sale_df['Previous_day_sale']
chennai_sale_df
```
